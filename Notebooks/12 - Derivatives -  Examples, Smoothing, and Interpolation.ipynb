#### Step-by-step forward differences example

```python
# Demonstration example 1:
#   Calculate the forward difference (FD) at one location (x[0])

import numpy as np
import matplotlib.pyplot as plt

# Define a function and sample it in x
x_lo = 0
x_hi = 6
n_step = 6
step_size = (x_hi+1 - x_lo) / n_step 
print("step_size", step_size)

x  = np.linspace(x_lo, x_hi, num=n_step, endpoint=False)
y  = 5*np.cos(x)

print("Point 0   x =",x[0]," y =",y[0])
print("Point 1   x =",x[1]," y =",y[1])

# Calculate the forward difference (FD) at one location (x[0]) 
forward_difference_0 = (y[1]-y[0])/step_size
print("Forward difference for location ",x[0]," =",forward_difference_0 )

# Plot the function
plt.plot(x,y,'o-',label='$y$')

# Plot just one line between x[0] and x[1] to demonstrate
#   the line used to calculate the slope for the FD method
plt.plot([x[0],x[1]], [y[0],y[1]], linestyle='-', marker='.', color='tab:orange', linewidth=3)

# Plot the FD at point x[0]
plt.plot(x[0], forward_difference_0, marker='o',color='tab:orange')
plt.grid(True)
plt.show()
```

```python
# Example - Derivative of a sampled function

import numpy as np
import matplotlib.pyplot as plt

# Define a function y = cos((-x^2)/3) and sample it in x
x_lo = 0
x_hi = 5
n_step = 1000
step_size = (x_hi - x_lo) / n_step 

x  = np.linspace(x_lo, x_hi, num=n_step, endpoint=True)
y  = np.cos(-(x**2)/3.0)

# Caculate the derivative of this sampled function
dy_fd = np.empty([x.size],float)
dy_bd = np.empty([x.size],float)
dy_cd = np.empty([x.size],float)

for ix in range(x.size):        
    value_current  = y[ix]

    # Calculate forward difference everywhere
    if ix<x.size-1:
        value_forward  = y[ix+1]
        dy_fd[ix] = (value_forward - value_current)/(step_size)

    # Calculate backward difference everywhere
    if ix>0:
        value_backward = y[ix-1]
        dy_bd[ix] =(value_current - value_backward)/(step_size)
    
    # Avoid the edges for calculating the central difference
    if ix>0 and ix<x.size-1:
        value_forward  = y[ix+1]
        value_backward = y[ix-1]
        dy_cd[ix] = (value_forward - value_backward)/(2*step_size)

               
plt.figure(figsize=(16,7))
plt.plot(x,y,'o-',label='$y$')
plt.plot( x[0:x.size-1], dy_fd[0:x.size-1],'o-',label='$y\'$ - Forward Difference')
plt.plot(         x[1:],         dy_bd[1:],'o-',label='$y\'$ - Backward Difference')
plt.plot( x[1:x.size-1], dy_cd[1:x.size-1],'o-',label='$y\'$ - Central Difference')
plt.grid(True)
plt.legend(loc='upper left', fontsize='xx-large')

plt.show()
```

```python
# Example - Derivative of a sampled function

import numpy as np
import matplotlib.pyplot as plt

# Define a function y = cos((-x^2)/3) and sample it at 100 points in x
x_lo = 0
x_hi = 5
n_step = 100
step_size = (x_hi - x_lo) / n_step 

x  = np.linspace(x_lo, x_hi, num=n_step, endpoint=True)
y  = np.cos(-(x**2)/3.0)

# Caculate the derivative of this sampled function
dy = np.empty([x.size],float)

for ix in range(x.size):        
    value_current  = y[ix]

    # Calculate forward difference on low edges
    if ix==0:
        value_forward  = y[ix+1]
        dy[ix] = (value_forward - value_current)/(step_size)

    # Calculate backward difference on high edges
    elif ix==x.size-1:
        value_backward = y[ix-1]
        dy[ix] =(value_current - value_backward)/(step_size)
    
    # Avoid the edges for calculating the central difference
    else:
        value_forward  = y[ix+1]
        value_backward = y[ix-1]
        dy[ix] = (value_forward - value_backward)/(2*step_size)

plt.figure(figsize=(16,7))
plt.plot(x,y,'o-',label='$y$')
plt.plot(x,dy,'o-',label='$y\'$')
plt.grid(True)
plt.legend(loc='upper left', fontsize='xx-large')

plt.show()
```

### How do we take the derivative of noisy data?

```python
# Noisy data

import numpy as np
import matplotlib.pyplot as plt

# Define y = f(x) = sin(x) + random noise
x = np.linspace(0,5*np.pi,200)
y = np.sin(x) + np.random.random(200) * 0.8

# Plot y
plt.figure(figsize=(16,7))
plt.plot(x,y)
plt.title('Data (sin(x)+ noise)', fontsize=20)
plt.show()

# Take a derivative
h = 5*np.pi/200
ysize =  y.shape[0]

dy = np.empty(ysize,float)

#   In the interior, use the central differences technique (2h)
dy[1:ysize-1]=(y[2:ysize]-y[0:ysize-2])/(2*h)

#   On each edge use forward/backword differences
dy[0] = (y[1]-y[0])/h
dy[ysize-1] = (y[ysize-1]-y[ysize-2])/h

plt.figure(figsize=(16,7))
plt.plot(x,dy, color='green')
plt.title('Derivative', fontsize=20)

model = np.cos(x) 
plt.plot(x,model, color='lightblue')

plt.show()
```

```python
# Smooth with a moving average 

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

x = np.linspace(0,5*np.pi,200)
y = np.sin(x) + np.random.random(200) * 0.8


N=len(x)
r=6
window_size = 2*r+1
points = N+window_size

all_window_sums   = []
for idx, j in enumerate(y[r:-r]):
    window_sum = y[idx:idx+window_size].sum()
    all_window_sums.append(window_sum)

a = np.array(all_window_sums,float)
av = a/window_size


plt.figure(figsize=(16,7))
plt.plot(x,y)
plt.plot(x[r:-r],av, color='red', linewidth=3, label='Moving average')
#plt.title('Smoothed data', fontsize=20)
plt.show()

# Take a derivative
h = 5*np.pi/200
ysize =  av.shape[0]
#print("ysize",ysize)
dy = np.empty(ysize,float)

#   In the interior, use the central differences technique (2h)
dy[1:ysize-1]=(av[2:ysize]-av[0:ysize-2])/(2*h)

#   On each edge use forward/backword differences
dy[0] = (av[1]-av[0])/h
dy[ysize-1] = (av[ysize-1]-av[ysize-2])/h

plt.figure(figsize=(16,7))
plt.plot(x[r:-r],dy, color='green')
plt.title('Derivative of moving average', fontsize=20)

model = np.cos(x) 
plt.plot(x,model, color='lightblue')

plt.show()
```

```python
# Smooth with a Savitzky-Golay filter

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

x = np.linspace(0,5*np.pi,200)
y = np.sin(x) + np.random.random(200) * 0.8

#  Savitzky-Golay filter
y_smooth1 = savgol_filter(y, 51, 3)  # savgol_filter(x, window_length, polyorder, deriv=0, delta=1.0, axis=-1, mode='interp', cval=0.0)[source]
y_smooth2 = savgol_filter(y, 7, 3)  # savgol_filter(x, window_length, polyorder, deriv=0, delta=1.0, axis=-1, mode='interp', cval=0.0)[source]

plt.figure(figsize=(16,7))
plt.plot(x,y)
plt.plot(x,y_smooth1, color='red')
plt.plot(x,y_smooth2, color='orange')
plt.title('Smoothed data', fontsize=20)
plt.show()

# Take a derivative
h = 5*np.pi/200
ysize =  y_smooth1.shape[0]
#print("ysize",ysize)
dy = np.empty(ysize,float)

#   In the interior, use the central differences technique (2h)
dy[1:ysize-1]=(y_smooth1[2:ysize]-y_smooth1[0:ysize-2])/(2*h)

#   On each edge use forward/backword differences
dy[0] = (y_smooth1[1]-y_smooth1[0])/h
dy[ysize-1] = (y_smooth1[ysize-1]-y_smooth1[ysize-2])/h

plt.figure(figsize=(16,7))
plt.plot(x,dy, color='green')
plt.title('Derivative of smoothed data', fontsize=20)

model = np.cos(x) 
plt.plot(x,model, color='lightblue')

plt.show()
```
