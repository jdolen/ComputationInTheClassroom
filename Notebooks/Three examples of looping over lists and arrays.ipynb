```python
# Simple loop over the values stored in a list
#   You can always use an integer to count iterations

import numpy as np
import matplotlib.pyplot as plt

# Read data
data = np.loadtxt("../Data/sunspots.txt",float)
list_months = data[:,0] # number of months since January 1749
list_Nspots = data[:,1] # number of sunspots observed per month

count = 0
for i in list_Nspots:
    print(count,list_months[count], i, list_Nspots[count])
    count+=1
```

```python
# Utilize enumerate to return both the value at
#   each entry in the list and the iteration count

import numpy as np
import matplotlib.pyplot as plt

# Read data
data = np.loadtxt("../Data/sunspots.txt",float)
list_months = data[:,0] # number of months since January 1749
list_Nspots = data[:,1] # number of sunspots observed per month
N=len(list_months)

for count, nspots in enumerate(list_Nspots[4:5]):
    print(count,nspots,list_months[count],list_Nspots[count])
```

```python
# To do the same thing with numpy arrays 
#   we can also use nditer
#   to return the value and the iteration

import numpy as np
import matplotlib.pyplot as plt

# Read data
data = np.loadtxt("../Data/sunspots.txt",float)
list_months = data[:,0] # number of months since January 1749
list_Nspots = data[:,1] # number of sunspots observed per month
array_months = np.array(list_months)
array_Nspots = np.array(list_Nspots)

it = np.nditer(array_Nspots, flags=['f_index']) # track the index
for value in it:
    print("value",value,"index",it.index )
```
